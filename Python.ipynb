```python
Write a python program to print the current year, month and day.
```

```python
# importing date class from datetime module 

from datetime import date 

  
# creating the date object of today's date 

todays_date = date.today() 

  
# printing todays date 

print("Current date: ", todays_date) 

  
# fetching the current year, month and day of today 

print("Current year:", todays_date.year) 

print("Current month:", todays_date.month) 

print("Current day:", todays_date.day)
```

```python
Write a Python program to construct the following pattern, using a
nested for loop.
*
**
***
****
*****
```

```python
n=5;
for i in range(n):
    for j in range(i):
        print ('* ', end="")
    print('')

for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```

```python
 Display all duplicate items from a list given:
10, 20, 60, 30, 20, 40, 30, 60, 70, 80
```

```python

def Repeat(x):

    _size = len(x)

    repeated = []

    for i in range(_size):

        k = i + 1

        for j in range(k, _size):

            if x[i] == x[j] and x[i] not in repeated:

                repeated.append(x[i])

    return repeated
 
# Driver Code

list1 = [10,20,60,30,20,40,30,60,70,80]


print (Repeat(list1))
```

```python
Write a python program to print the numbers from a given number n
till 0 using recursion
```

```python

n=int(input("n"))
def countdown(n):
    print(n)
    if n == 0:
        return 0
    return countdown(n - 1)
```

```python
countdown(8)
```
